## Activity 1.01 Handling Lists

> These are the steps for completing this activity:
> 1. Create a **list** of 100 random numbers.
> 2. Create a new **list** from this random **list**, with numbers that are divisible by **3**.
> 3. Calculate the length of these two lists and store the difference in a new variable.
> 4. Using a loop, perform steps 1, 2, and 3, and find the difference variable 10 times.
> 5. Find the arithmetic mean of these **10** difference values.
> 
> The output (will vary with each run) should look similar to this:
> 
> 66.3

### Step 1: Create a list of 100 random numbers

```python
# Reinitialize the random number list for the next activity
random_number_list = [
    # Randint generates a random integer between the two specified values for each iteration
    random.randint(0, 100) for x in range(0, 100)
]
random_number_list
```

### Step 2: Create a new list with numbers divisible by 3

```python
divisible_by_3_list = [
    # For loop to check each number in the random number list if it is divisible by 3
    num for num in random_number_list if num % 3 == 0
]

divisible_by_3_list
```

### Step 3: Calculate the length difference between the two lists

```python
# len function calculates the length of a list
length_difference = len(random_number_list) - len(divisible_by_3_list)

length_difference
```

### Step 4: Repeat steps 1-3 ten times and store the differences

```python
# Initialize a list to store the difference values
difference_values = []

for _ in range(10):
    # Step 1: Create a new list of 100 random numbers
    random_number_list = [random.randint(0, 100) for x in range(0, 100)]
    
    # Step 2: Create a new list with numbers divisible by 3
    divisible_by_3_list = [num for num in random_number_list if num % 3 == 0]
    
    # Step 3: Calculate the length difference between the two lists
    length_difference = len(random_number_list) - len(divisible_by_3_list)
    
    # Store the difference value
    difference_values.append(length_difference)

# Print the results of the differences
difference_values
```

### Step 5: Calculate the arithmetic mean of the differences

```python
# Calculate the arithmetic mean of the differences
arithmetic_mean = sum(difference_values) / len(difference_values)

arithmetic_mean
```

### Step 3: Remove all new lines and symbols using the replace method

Just removing symbols and new lines below. This will require further cleaning outside of a for loop.

```python
# Create lists of patterns to replace
replace_patterns_w_space = ['\n','{','}','-','']

# Optional variable to assist with removing multiple spaces
replace_patterns_no_space = [',', '.', '"', '', '', '?', '!', ':',';']

# Initialize cleaned_text with the original multiline_text
cleaned_multiline_text = multiline_text

# Remove new lines and symbols using replace method in a for loop
for word in multiline_text:
    # Check each word against the patterns to replace
    if word in replace_patterns_w_space:
        cleaned_multiline_text = cleaned_multiline_text.replace(word, ' ')

    # Second condition to remove symbols without adding space
    elif word in replace_patterns_no_space:
        cleaned_multiline_text = cleaned_multiline_text.replace(word, '')
    else:
        # If the character is not in either list, leave it as is
        continue

# Final cleanup to remove any double spaces and leading/trailing spaces
cleaned_multiline_text = cleaned_multiline_text.replace('  ', ' ').strip()

cleaned_multiline_text
```

---

Just for my own sanity, I am doing another method below to account for missed numeric characters that don't make grammatical sense. I want to keep only letters, digits, and whitespace characters. Whitespace characters include spaces, tabs, and newlines but, spaces should typically be 1 space between words.

I can import re module to use regular expressions for this task. It provides a substitution method that can replace unwanted characters based on a pattern.

The pattern `r'[^A-Za-z0-9\s]'` matches any character that is not a letter (uppercase or lowercase) [`A-Za-z`], digit [`0-9`], or whitespace [`\s`]. I will replace these characters with a space and strip leading and trailing whitespace.

```python
import re

# Wiping slate clean so no one thinks I am cheating!
cleaned_multiline_text = None

# Keep only letters, digits, and whitespace characters
cleaned_multiline_text = re.sub(r'[^A-Za-z0-9\s]', ' ', multiline_text)

# Normalize whitespace to single spaces and strip leading/trailing spaces
cleaned_multiline_text = re.sub(r'\s+', ' ', cleaned_multiline_text).strip()
```

### Step 6: Count the number of times the unique words has appeared in the list using the **key** and **value** in **dict**

Will need to convert the unique words list into a dictionary with keys as unique words and values initialized to zero or left empty.

Will need to create a for loop to iterate through the unique words list and count occurrences in the words list.

```python
# Create a dictionary with unique words as keys
unique_words_as_dict = dict.fromkeys(unique_words_list)
unique_words_as_dict
```

```python
# Iterate through the words list to count occurrences from dictionary
for word in words_list:
    # Checks the first time the word is encountered
    if unique_words_as_dict[word] is None:
        # Sets the count to 1 since it's the first occurrence
        unique_words_as_dict[word] = 1
    else:
        # Increments the count for subsequent occurrences
        unique_words_as_dict[word] += 1

unique_words_as_dict
```

`assert` is used to verify that a condition is true. If the condition is false, it raises an `AssertionError`. In this case, we want to assert that each element from the permutations iterator is indeed a tuple.

### Step 4: Use dropwhile with a lambda to drop leading zeros and cast output to a list

Using a for loop, I'll print out the list of possible numbers using permutations as my range. Using `dropwhile` I can pass a **lambda** function that drops leading zeros from each tuple. The **lambda** function will only check the first element of the tuple because `dropwhile` checks elements from the start of the iterable and then stops and goes to the next tuple.

```python
for number_tuple in permutations(range(3)):
    # Print after dropping any leading zeros
    print(list(dropwhile(
        # x represents the first element of each tuple
        lambda x: x == 0, number_tuple)
        )
    )  
```

### Step 5: Check the actual type that dropwhile returns

Creating a function to convert a list of digits into a whole number will need a **lambda** function. `reduce` from the `functools` module can be used to apply a rolling computation to sequential pairs of values in a list. In this case, it will help in constructing the whole number from its individual digits. Coming from the documentation from `reduce`: 

`reduce`(**lambda** x, y: x+y, [1, 2, 3, 4, 5]) calculates ((((1 + 2) + 3) + 4) + 5)

I can change the caluclation of x and y to multiply x by 10 and add y to it. This will effectively shift the digits to the left and add the new digit at the end, constructing the whole number as we iterate through the list of digits.

```python
from functools import reduce

def to_number(number_stack):
    # Multiply the first element by 10 and add the second element
    return reduce(lambda x, y: x * 10 + y, number_stack)

# Recreate the prior for loop with to_number function
for number_tuple in permutations(range(3)):
    number_stack = list(dropwhile(
        # x represents the first element of each tuple
        lambda x: x == 0, number_tuple)
        )
    print(float(to_number(number_stack)))
```
