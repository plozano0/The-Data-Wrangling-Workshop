## Step 13: Use `subset` and `groupby` to find the outliers.

```python
# Use subset and groupby to find the outliers by 
adult_income_df
```

I'll decide that the outliers I will look for are those individuals whose age is above the 75th percentile and below the 25th percentile for each occupation. This means I will identify individuals who are significantly older or younger than the typical age range for their respective occupations.

```python
# Find the outliers by age within each occupation
def find_age_outliers(group):
    Q1 = group['age'].quantile(0.25)
    Q3 = group['age'].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = group[(group['age'] < lower_bound) | (group['age'] > upper_bound)]
    return outliers

outliers_df = adult_income_df.groupby('occupation').apply(find_age_outliers).reset_index(drop=True)
outliers_df
```

## Step 14: Plot the outlier values on a bar chart.

I will plot the outliers using a horizontal bar plot for each occupation to help visualize the age distribution and highlight the outliers.

```python
# Horizontal Bar plot to visualize the number of outliers in each occupation
outlier_counts = outliers_df['occupation'].value_counts().sort_values(ascending=True)
plt.figure(figsize=(10, 6))
outlier_counts.plot(
    kind='barh', 
    color='skyblue', 
    edgecolor='black'
)
plt.title('Number of Age Outliers by Occupation')
plt.xlabel('Number of Outliers')
plt.ylabel('Occupation')

# gca = get current axis
ax = plt.gca()
# Provide labels for each bar using bar_label
# axis containers[0] refers to the first set of bars in the plot
ax.bar_label(ax.containers[0], label_type='edge') # <-- label_type can be 'center', 'edge', or 'inside'
plt.show()
```

## Step 15: Merge the two DataFrames using common keys to drop duplicate values.

Using the same `adult_income_df` DataFrame from earlier steps, I will create a new DataFrame called `education_occupation_df` that contains only the **education** and **occupation** columns. Then, I will merge this new DataFrame with another DataFrame called `age_occupation_df`, which contains the **age** and **occupation** columns. After merging, I will drop any duplicate values to ensure that each occupation is represented only once in the final DataFrame.

```python
# Create a new DataFrame with only education and occupation columns
education_occupation_df = adult_income_df[['education',
            # Add sample data to it with replacement (to allow duplicates)
           'occupation']].sample(5,random_state=101, replace=True)
education_occupation_df.head()
```

## Step 1: Create two series as shown below with the given index values.

> Series 1 = 7.3, -2.5, 3.4, 1.5
>
> i. Index = 'a', 'b', 'c', 'd', 'e'
>
> Series 2 = 2.1, 3.6, -1.5, 4, 3.1
>

```python
# Series 1 = 7.3, -2.5, 3.4, 1.5
series1 = pd.Series([7.3, -2.5, 3.4, 1.5], index=['a', 'b', 'c', 'd'])
series1
```

```python
series2 = pd.Series([2.1, 3.6, -1.5, 4, 3.1], index=['a', 'c', 'e', 'f', 'g'])
series2
```

## Step 2: Add series 1 and series 2 together and print the results

```python
# Add the two series together by use of the add() method
series_sum = series1.add(
    series2,
    # Fill missing values with 0
    fill_value=0)

# Print the result
series_sum
```

## Step 3: Subtract series 1 from series 2 and print the results

```python
# Subtract series 1 from series 2 and print the results
series_diff = series1.subtract(
    series2,
    # Fill missing values with 0
    fill_value=0
)

# Print the result
series_diff
```

Alternatively, I could just use the `+` and `-` operators to add and subtract the two series directly. However, using the `add()` and `subtract()` methods allows me to specify how to handle missing values, which is important in this case since the two series have different indices. By using the `fill_value=0` parameter, I ensure that any missing values are treated as zero during the arithmetic operations, resulting in a more accurate outcome.
